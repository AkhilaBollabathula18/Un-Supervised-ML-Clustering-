```python
x=df.drop(["country"],axis=1)
```

```python
x.head()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_scale=sc.fit_transform(x)
x_scale
```

```python
x_scale_df=pd.DataFrame(x_scale,columns=x.columns)
x_scale_df.head()
```

```python
from scipy.cluster.hierarchy import dendrogram
import scipy.cluster.hierarchy as shc
```

```python
plt.figure(figsize=(10,7))
plt.title("Country dendrogram")
dend=shc.dendrogram(shc.linkage(x_scale_df,method="ward"))
```

```python
from scipy.cluster.hierarchy import linkage
margin=linkage(x_scale_df,method="complete",metric="euclidean")
plt.title("Complete linkage of data")
dendrogram(margin)
plt.show()
```

```python
from scipy.cluster.hierarchy import cut_tree
cluster_tree=cut_tree(margin,n_clusters=2).reshape(-1,1)
cluster_tree
```

```python
x_scale_df["Hierarcal clustering"]=cluster_tree
x_scale_df.head(50)
```

```python
x_scale_df["Hierarcal clustering"].value_counts()
```
